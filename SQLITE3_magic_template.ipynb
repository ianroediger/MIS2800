```python
%%sql

CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    category TEXT NOT NULL,
    price REAL NOT NULL
) STRICT;

INSERT INTO products (name, category, price) VALUES
('Superbowl Jersey', 'Clothing', 120.00),
('Headphones', 'Electronics', 85.50),
('Mouse', 'Electronics', 45.99);
```

```python
%%sql

SELECT * FROM products;

SELECT * FROM products
WHERE price > 50;
```

```python
%%sql

UPDATE products
SET price = 99.99
WHERE name = 'Superbowl Jersey';

SELECT * FROM products
WHERE name = 'Superbowl Jersey';

DELETE FROM products
WHERE product_id = 1;

SELECT * FROM products;
```
